```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
from time import time
from sklearn.utils import check_random_state
from sklearn import manifold, datasets
```

```python

# Next line to silence pyflakes.
Axes3D

# Variables for manifold learning.
n_points = 25
X, colors = datasets.samples_generator.make_s_curve(n_points, random_state=0)

n_components = 2

x=X[:, 0]
y=X[:, 1]
z=X[:, 2]
# Plot our dataset.
fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(251, projection='3d')
ax.scatter(x, y, z, c=colors, cmap=plt.cm.Spectral)
ax.view_init(4, -72)

sphere_data = np.array([x, y, z]).T
plt.show()
# Perform Locally Linear Embedding Manifold learning
neilist=[5,10, 20]
cnt=1
for numnei in neilist:
    fig=plt.figure()
    trans_data = manifold.LocallyLinearEmbedding(numnei, 2,method='standard').fit_transform(sphere_data).T

    ax = fig.add_subplot(3,3,cnt)
    plt.scatter(trans_data[0], trans_data[1], c=colors, cmap=plt.cm.rainbow)
    plt.title("%s (%d)" % ('LLE', numnei))
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    plt.axis('tight')


    trans_data = manifold.Isomap(numnei, n_components=2).fit_transform(sphere_data).T
    ax = fig.add_subplot(3,3,cnt+1)
    plt.scatter(trans_data[0], trans_data[1], c=colors, cmap=plt.cm.rainbow)
    plt.title("%s (%d)" % ('Isomap', numnei))
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    plt.axis('tight')


    mds = manifold.MDS(2, max_iter=100, n_init=1)
    trans_data = mds.fit_transform(sphere_data).T
    ax = fig.add_subplot(3,3,cnt+2)
    plt.scatter(trans_data[0], trans_data[1], c=colors, cmap=plt.cm.rainbow)
    plt.title("MDS (%d)" %numnei)
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
    plt.axis('tight')

    cnt+=3

    plt.show()
```
