```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
import cmath
from sklearn.datasets import make_moons,make_classification
from sklearn.lda import LDA
from sklearn.decomposition import PCA 
```

```python
data, lbl = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
data += 2 * rng.uniform(size=data.shape)
```

```python
clf = LDA()
clf.fit(data, lbl)
score = clf.score(data, lbl)
x_min, x_max = data[:, 0].min() - .5, data[:, 0].max() + .5
y_min, y_max = data[:, 1].min() - .5, data[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.025), np.arange(y_min, y_max, 0.025))

    
if hasattr(clf, "decision_function"):
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
else:
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z)
plt.contour(xx, yy, Z, [0.5], linewidths=2., colors='k')
    
plt.scatter(data[:, 0], data[:, 1], c=lbl)
plt.show()
```
